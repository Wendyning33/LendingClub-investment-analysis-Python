```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
%matplotlib inline
import numpy as np
import statsmodels.api as sm
import math
import heapq as hq
```

```python
lc=pd.read_csv("lc2final.csv")
```

```python
x=lc[['B', 'C', 'D', 'E', 'F', 'G', 'dti', 'fico_avg', 'emp_length', 'revol_bal', 'term_num']]
y=lc['default']
x=sm.tools.tools.add_constant(x)
logit=sm.Logit(y,x)

result=logit.fit()
result.summary()
```

```python
from sklearn.neighbors import KNeighborsClassifier
x=lc[['B', 'C', 'D', 'E', 'F', 'G', 'dti', 'fico_avg', 'emp_length', 'revol_bal', 'term_num']]
y=lc['default']
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(x, y)
```

```python
neigh.predict(x.loc[0:100])
```

```python
neigh.predict_proba(x.loc[0:10])
```
